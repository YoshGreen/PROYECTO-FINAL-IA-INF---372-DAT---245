```python
import pandas as pd
import numpy as np
```

#¿Qué hace?
*   Carga el dataset desde el archivo CSV.
*   Muestra las primeras filas para verificar que los datos se leyeron correctamente.

Permite validar:

```python
path = "/content/data.csv"
df = pd.read_csv(path)

print("Shape:", df.shape)
df.head()
```

#¿Qué hace?
Cuenta cuántos valores nulos existen por columna.
#¿Por qué es importante?
Porque:
*   los modelos de IA no aceptan valores faltantes
*   necesitamos saber si debemos imputar o no.

```python
print("Valores faltantes por columna:")
df.isnull().sum()
```

#¿Qué hace?
*   Elimina id porque no aporta información predictiva.
*   Convierte la variable objetivo diagnosis a formato numérico.
#¿Por qué es importante?
*   Los modelos trabajan solo con números.
*   id puede introducir ruido si se mantiene.

```python
# Eliminar ID (no aporta al modelo)
df = df.drop("id", axis=1)

# Convertir diagnosis B/M -> 0/1
df["diagnosis"] = df["diagnosis"].map({"B":0, "M":1})

df.head()
```

#¿Qué hace?
*   Fuerza todas las columnas predictoras a tipo numérico.
*   Convierte valores mal interpretados en NaN.
#¿Por qué es importante?
Porque algunos valores pueden:
*   venir como texto
*   tener caracteres invisibles
*   causar errores en escalado y SMOTE.





```python
for col in df.columns:
    if col != "diagnosis":
        df[col] = pd.to_numeric(df[col], errors="coerce")

df.dtypes
```

#¿Qué hace?
Reemplaza valores faltantes por la mediana de cada columna.
#¿Por qué mediana y no media?
*   Es más robusta frente a valores extremos.
*   Muy usada en datasets biomédicos.

```python
# Imputación con mediana
df = df.fillna(df.median(numeric_only=True))

print("Faltantes después de imputación:")
df.isnull().sum().sum()
```

#¿Qué hace?
Separa:
*   X → variables predictoras
*   y → variable objetivo
#¿Por qué es importante?
Todos los modelos de scikit-learn requieren esta separación.

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]

print("X:", X.shape, "y:", y.shape)
```

#¿Qué hace?
Elimina columnas con un solo valor en todas las filas.
#¿Por qué es clave?
*   No aportan información.
*   Generan NaN al estandarizar (división entre cero).
*   Rompen SMOTE.

```python
zero_var_cols = [c for c in X.columns if X[c].nunique() <= 1]
print("Columnas con varianza cero:", zero_var_cols)

X = X.loc[:, X.nunique() > 1]
print("X después de eliminar varianza cero:", X.shape)
```

#¿Qué hace?
Escala las variables para que tengan:
*   media = 0
*   desviación estándar = 1
#¿Por qué es importante?
*   Las variables tienen rangos muy distintos.
*   Modelos como KNN, SVM y Regresión Logística lo requieren.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# A) Normalización (0-1) - opcional
minmax = MinMaxScaler()
X_minmax = pd.DataFrame(minmax.fit_transform(X), columns=X.columns)

# B) Estandarización (Z-score) - recomendada
scaler = StandardScaler()
X_zscore = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("NaN después de escalar:", X_zscore.isnull().sum().sum())
print("Inf después de escalar:", np.isinf(X_zscore).sum().sum())

X_zscore.head()
```

#¿Qué hace?
*   Crea nuevas variables a partir de relaciones geométricas.
#¿Por qué es útil?
*   Las proporciones suelen ser más informativas que valores absolutos.
*   Mejora la capacidad predictiva del modelo.

```python
df_fe = X_zscore.copy()

df_fe["radius_per_area"] = (X["radius_mean"] / (X["area_mean"] + 1e-6)).values
df_fe["compactness_ratio"] = (X["compactness_mean"] / (X["concavity_mean"] + 1e-6)).values

df_fe.head()
```

#¿Qué hace?
Divide los datos en:
*   70% entrenamiento
*   30% prueba

Sin aplicar SMOTE

Muestra la distribución real de clases en el conjunto de entrenamiento.

Aplica SMOTE solo al conjunto de entrenamiento.

Muestra la nueva distribución balanceada después de SMOTE.
#¿Por qué?
*   Permite ver explícitamente el desbalance del dataset.
*   Demuestra que SMOTE no se aplica al conjunto de prueba (evita data leakage).
*   Justifica metodológicamente el uso de SMOTE.
*   Es evidencia visual para el informe y la defensa.

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd

# Split SIN SMOTE
X_train_ns, X_test_ns, y_train_ns, y_test_ns = train_test_split(
    X_zscore, y, test_size=0.3, random_state=42, stratify=y
)


print("Distribución ORIGINAL en entrenamiento:")
print(y_train_ns.value_counts())

# Aplicar SMOTE SOLO al entrenamiento
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train_ns, y_train_ns)

print("\nDistribución BALANCEADA con SMOTE (solo entrenamiento):")
print(pd.Series(y_train_sm).value_counts())
```

#¿Qué hace?
1. Usa el mismo split que la celda anterior.
2. Entrena el mismo modelo en dos escenarios:
  *   sin SMOTE
  *   con SMOTE
3. Evalúa ambos modelos sobre el mismo conjunto de prueba real.
4. Calcula métricas:
  *   Accuracy
  *   Precision
  *   Recall
  *   F1-score
  *   Matriz de confusión
#¿Por qué es importante?
*   Permite comparar desempeño real, no solo balance de datos.
*   Muestra si SMOTE:
*   mejora recall (detección de malignos),
*   empeora precision,
*   cambia el tipo de errores.
*   Es fundamental en problemas médicos, donde los falsos negativos importan más.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Modelo
model = DecisionTreeClassifier(random_state=42)

# ===== ESCENARIO 1: SIN SMOTE =====
model.fit(X_train_ns, y_train_ns)
pred_ns = model.predict(X_test_ns)

print("====== SIN SMOTE ======")
print("Accuracy :", accuracy_score(y_test_ns, pred_ns))
print("Precision:", precision_score(y_test_ns, pred_ns))
print("Recall   :", recall_score(y_test_ns, pred_ns))
print("F1-score :", f1_score(y_test_ns, pred_ns))
print("Matriz de Confusión:\n", confusion_matrix(y_test_ns, pred_ns))

# ===== ESCENARIO 2: CON SMOTE =====
model.fit(X_train_sm, y_train_sm)
pred_sm = model.predict(X_test_ns)

print("\n====== CON SMOTE ======")
print("Accuracy :", accuracy_score(y_test_ns, pred_sm))
print("Precision:", precision_score(y_test_ns, pred_sm))
print("Recall   :", recall_score(y_test_ns, pred_sm))
print("F1-score :", f1_score(y_test_ns, pred_sm))
print("Matriz de Confusión:\n", confusion_matrix(y_test_ns, pred_sm))
```

#¿Qué hace?
Esta celda define y evalúa distintos algoritmos de clasificación supervisada:
*   Árbol de Decisión
*   K-Nearest Neighbors (KNN)
*   Regresión Logística
*   Naive Bayes
*   Support Vector Machine (SVM)

Cada modelo se entrena utilizando el conjunto de entrenamiento original sin aplicar SMOTE y se evalúa sobre el mismo conjunto de prueba. Para cada modelo se calculan métricas de desempeño como:
*   Accuracy
*   Precision
*   Recall
*   F1-score
*   Matriz de confusión

Finalmente, los resultados se almacenan en una tabla comparativa que permite contrastar el desempeño de todos los modelos evaluados.
#¿Por qué?
*   Permite comparar distintos enfoques de clasificación bajo las mismas condiciones experimentales.
*   Garantiza una evaluación justa, ya que todos los modelos usan el mismo conjunto de entrenamiento y prueba.
*   Proporciona una línea base realista del desempeño de los modelos utilizando datos reales, sin técnicas de balanceo artificial.
*   Facilita la selección del modelo más adecuado para el problema de clasificación médica.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd

# Definición de modelos
modelos = {
    "Árbol de Decisión": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Regresión Logística": LogisticRegression(max_iter=2000),
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(kernel="linear")
}

resultados = []

for nombre, modelo in modelos.items():
    # Entrenamiento SIN SMOTE (datos reales)
    modelo.fit(X_train_ns, y_train_ns)
    pred = modelo.predict(X_test_ns)

    # Métricas
    acc = accuracy_score(y_test_ns, pred)
    prec = precision_score(y_test_ns, pred)
    rec = recall_score(y_test_ns, pred)
    f1 = f1_score(y_test_ns, pred)
    cm = confusion_matrix(y_test_ns, pred)

    resultados.append([nombre, acc, prec, rec, f1])

    print("\n==============================")
    print("Modelo:", nombre)
    print("==============================")
    print("Matriz de Confusión:\n", cm)
    print(f"Accuracy:  {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall:    {rec:.4f}")
    print(f"F1-Score:  {f1:.4f}")

# Tabla comparativa final
df_resultados = pd.DataFrame(
    resultados,
    columns=["Modelo", "Accuracy", "Precision", "Recall", "F1"]
)

df_resultados
```

#¿Qué hace?
Esta celda define una función que evalúa el desempeño de un modelo de clasificación utilizando múltiples divisiones aleatorias del dataset.
Para cada iteración:
1. Se divide el dataset en entrenamiento y prueba según una proporción definida (por ejemplo 80/20 o 50/50).
2. Opcionalmente, se aplica la técnica SMOTE solo al conjunto de entrenamiento.
3. Se entrena un modelo SVM con kernel lineal.
4. Se evalúa el modelo sobre el conjunto de prueba y se calcula la exactitud (accuracy).

Este proceso se repite 100 veces y la función devuelve la mediana de las exactitudes obtenidas.

#¿Por qué?
*   Reduce la dependencia de un único split entrenamiento/prueba.
*   Permite evaluar la estabilidad y robustez del modelo frente a diferentes particiones de los datos.
*   El uso de la mediana proporciona una medida más robusta que el promedio, evitando la influencia de valores atípicos.
*   Cumple con los criterios de validación solicitados en el enunciado académico del proyecto.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE
import numpy as np

def evaluar_modelo_correcto(X, y, test_size, usar_smote=False, n_reps=100):
    accuracies = []

    for i in range(n_reps):
        # 1. Split primero (datos reales)
        X_train, X_test, y_train, y_test = train_test_split(
            X, y,
            test_size=test_size,
            random_state=i,
            stratify=y
        )

        # 2. SMOTE SOLO en entrenamiento
        if usar_smote:
            sm = SMOTE(random_state=i)
            X_train, y_train = sm.fit_resample(X_train, y_train)

        # 3. Entrenar modelo
        modelo = SVC(kernel="linear")
        modelo.fit(X_train, y_train)

        # 4. Evaluar
        y_pred = modelo.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        accuracies.append(acc)

    return np.median(accuracies)
```

#¿Qué hace?
Esta celda utiliza la función definida previamente para evaluar el modelo bajo distintos escenarios experimentales:
*   División 80% entrenamiento / 20% prueba sin SMOTE.
*   División 50% entrenamiento / 50% prueba sin SMOTE.
*   División 80% entrenamiento / 20% prueba aplicando SMOTE solo al entrenamiento.
*   División 50% entrenamiento / 50% prueba aplicando SMOTE solo al entrenamiento.
Para cada caso se calcula la mediana de la exactitud obtenida tras 100 iteraciones y se muestran los resultados por pantalla.
#¿Por qué?
*   Permite analizar cómo influye la proporción de datos de entrenamiento y prueba en el desempeño del modelo.
*   Evalúa el impacto del balanceo de clases mediante SMOTE en distintos escenarios.
*   Proporciona una validación más robusta y confiable del modelo, especialmente relevante en contextos médicos.
*   Ayuda a determinar si el uso de SMOTE mejora la estabilidad o el rendimiento del clasificador.

```python
# SIN SMOTE
med_80_20_no = evaluar_modelo_correcto(X_zscore, y, test_size=0.2, usar_smote=False)
med_50_50_no = evaluar_modelo_correcto(X_zscore, y, test_size=0.5, usar_smote=False)

# CON SMOTE (solo entrenamiento)
med_80_20_sm = evaluar_modelo_correcto(X_zscore, y, test_size=0.2, usar_smote=True)
med_50_50_sm = evaluar_modelo_correcto(X_zscore, y, test_size=0.5, usar_smote=True)

print("SIN SMOTE")
print("Mediana Accuracy 80/20:", med_80_20_no)
print("Mediana Accuracy 50/50:", med_50_50_no)

print("\nCON SMOTE")
print("Mediana Accuracy 80/20:", med_80_20_sm)
print("Mediana Accuracy 50/50:", med_50_50_sm)
```
