#**PROBLEMA DE LOS N CABALLOS RESUELTO POR EL METODO COMBINATORIO**


```python
def max_caballos(n):
    return (n*n + 1)//2


def ejemplo_caballos(n, k):
    tablero = []
    for fila in range(n):
        for col in range(n):
          # UTILIZA UN PATRON DE TABLERO DE AJEDREZ
            if (fila + col) % 2 == 0:
                tablero.append((fila, col))
    #TOMA SOLO LOS K CABALLOS
    return tablero[:k]


#FUNCION PARA MOSTRAR EL TABLERO
def mostrar_tablero(tablero, n):
    for fila in range(n):
        linea = ""
        for col in range(n):
            if (fila, col) in tablero:
                linea += "C "
            else:
                linea += ". "
        print(linea)
    print("\n")


n = 8
max_k = max_caballos(n)
print(f"Tablero de tamaño {n}x{n}")
print(f"Máximo número de caballos que se pueden colocar el el tablero: {max_k}")

k = 32

tablero = ejemplo_caballos(n, k)
print("EJEMPLO DEL TABLERO CON LAS :")
mostrar_tablero(tablero, n)
```

#***PROBLEMA DE LAS N REINAS CON Implementación funcional de Simulated Annealing***

```python
import random
import math

def contar_ataques(tablero):
    n = len(tablero)
    ataques = 0
    for i in range(n):
        for j in range(i+1, n):
            if tablero[i] == tablero[j]:
                ataques += 1
            if abs(tablero[i] - tablero[j]) == abs(i - j):
                ataques += 1
    return ataques

def vecino(tablero):
    n = len(tablero)
    nuevo = list(tablero)
    columna = random.randint(0, n-1)
    fila_nueva = random.randint(0, n-1)
    nuevo[columna] = fila_nueva
    return nuevo

def recocido_simulado(n, iteraciones=10000, temp_inicial=100, enfriamiento=0.99):
    tablero = [random.randint(0, n-1) for _ in range(n)]

    for _ in range(iteraciones):
        ataques = contar_ataques(tablero)
        if ataques == 0:
            return tablero, 0

        t = vecino(tablero)
        ataques_nuevo = contar_ataques(t)
        delta = ataques_nuevo - ataques

        if delta < 0 or random.random() < math.exp(-delta / temp_inicial):
            tablero = t

        temp_inicial *= enfriamiento

    return tablero, contar_ataques(tablero)

def mostrar_tablero(tablero):
    n = len(tablero)
    for fila in range(n):
        linea = ""
        for col in range(n):
            if tablero[col] == fila:
                linea += "Q "
            else:
                linea += ". "
        print(linea)
    print("\n")


sol, ataques = recocido_simulado(n, iteraciones=20000, temp_inicial=50, enfriamiento=0.995)

print(f"Vector de posiciones de las reinas: {sol}")
print(f"Número de ataques: {ataques}")
print("Tablero final:")
mostrar_tablero(sol)
```
